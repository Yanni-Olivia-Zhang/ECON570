```python
import pandas as pd
```

```python
twt = pd.read_csv('tweets_01-08-2021.csv', index_col = 'id')
```

```python
twt = twt[twt['isRetweet'] == 'f']
```

```python
twt_test = twt.head(5)
```

```python
d = dict()
for i in twt['text']:
    for word in i.lower().split(' '):
        if not word.startswith(('http','@','#')):
            if word.endswith((',','.','!','?',':')):
                word = word[:-1]
            if word not in d:
                d[word] = 0
            d[word] += 1
d
```
